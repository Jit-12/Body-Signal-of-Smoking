## Introduction

## Body signal of smoking
- Data Link: https://www.kaggle.com/datasets/kukuroo3/body-signal-of-smoking/data

#### About Dataset
    This dataset is a collection of basic health biological signal data.
    The goal is to determine the presence or absence of smoking through bio-signals.
    
#### Columns
   - Data Shape: (55692, 27)
      - ID : index
      - gender
      - age : 5-years gap
      - height(cm)
      - weight(kg)
      - waist(cm) : Waist circumference length
      - eyesight(left)
      - eyesight(right)
      - hearing(left)
      - hearing(right)
      - systolic : Blood pressure
      - relaxation : Blood pressure
      - fasting blood sugar
      - Cholesterol : total
      - triglyceride
      - HDL : cholesterol type
      - LDL : cholesterol type
      - hemoglobin
      - Urine protein
      - serum creatinine
      - AST : glutamic oxaloacetic transaminase type
      - ALT : glutamic oxaloacetic transaminase type
      - Gtp : γ-GTP
      - oral : Oral Examination status
      - dental caries
      - tartar : tartar status
      - smoking

## Building Machine Learning Models

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.model_selection import train_test_split,cross_val_score

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import  KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier

from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.metrics import roc_auc_score,roc_curve
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
df = pd.get_dummies(df,columns = ['gender','tartar'],drop_first=True)
```

```python
df = df.drop(columns=['height(cm)','weight(kg)','oral','HDL','LDL','ID'],axis=1)
```

```python
y = df["smoking"]
x = df.drop("smoking", axis = 1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.1, shuffle = True, random_state = 2)
```

```python
standard = StandardScaler()
standard.fit(x_train)
x_train_sd = standard.transform(x_train)
x_test_sd = standard.transform(x_test)
```

```python
rf_model = RandomForestClassifier(n_estimators = 500)
rf_model.fit(x_train_sd, y_train)
y_pred = rf_model.predict(x_test_sd)
accuracy_score(y_test, y_pred)
```

```python
XGB = XGBClassifier(n_estimators=2000)
XGB.fit(x_train_sd, y_train)
y_pred = XGB.predict(x_test_sd)
accuracy_score(y_test, y_pred)
```

```python
LR = LogisticRegression(max_iter=200)
LR.fit(x_train_sd, y_train)
y_pred = LR.predict(x_test_sd)
accuracy_score(y_test, y_pred)
```

```python
KNR = KNeighborsClassifier(n_neighbors=3)
KNR.fit(x_train_sd, y_train)
y_pred = KNR.predict(x_test_sd)
accuracy_score(y_test, y_pred)
```

```python
SVM=SVC()
SVM.fit(x_train_sd, y_train)
y_pred = SVM.predict(x_test_sd)
accuracy_score(y_test, y_pred)
```

## Model Selection Results
- Random Forest Classifier = 85.1 %
- XGB Classifier = 83.5 %
- SVC = 76.1%
- Logistic Regression = 73.8 %
- K Neighbors Classifier = 73.7 %

So , the best model Random Forest Classifier


## Random Forest

```python
rf_model = RandomForestClassifier(n_estimators = 500)
rf_model.fit(x_train_sd, y_train)
y_pred = rf_model.predict(x_test_sd)
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, rf_model.predict(x_test_sd), labels=rf_model.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=rf_model.classes_)
disp.plot()
plt.show()
```

```python
rf_roc_auc = roc_auc_score(y_test, rf_model.predict(x_test_sd))
fpr, tpr, thresholds = roc_curve(y_test, rf_model.predict_proba(x_test_sd)[:,1])

plt.figure()
plt.plot(fpr, tpr, label = 'AUC (area = %0.2f)' % rf_roc_auc)
plt.plot([0, 1], [0, 1],'g--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC')
plt.show()
```

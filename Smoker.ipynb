## Introduction

### Body signal of smoking
- Data Link: https://www.kaggle.com/datasets/kukuroo3/body-signal-of-smoking/data

#### About Dataset
    This dataset is a collection of basic health biological signal data.
    The goal is to determine the presence or absence of smoking through bio-signals.
    
#### Columns
   - Data Shape: (55692, 27)
      - ID : index
      - gender
      - age : 5-years gap
      - height(cm)
      - weight(kg)
      - waist(cm) : Waist circumference length
      - eyesight(left)
      - eyesight(right)
      - hearing(left)
      - hearing(right)
      - systolic : Blood pressure
      - relaxation : Blood pressure
      - fasting blood sugar
      - Cholesterol : total
      - triglyceride
      - HDL : cholesterol type
      - LDL : cholesterol type
      - hemoglobin
      - Urine protein
      - serum creatinine
      - AST : glutamic oxaloacetic transaminase type
      - ALT : glutamic oxaloacetic transaminase type
      - Gtp : γ-GTP
      - oral : Oral Examination status
      - dental caries
      - tartar : tartar status
      - smoking

### Visualizations

```python
plt.figure(figsize = [22, 12])
sns.heatmap(df.corr(), annot = True, square = False, linewidths = 2,cmap = 'inferno');
```

```python
df.hist(figsize = (20, 20), bins = 12, legend = False)
plt.show()
```

```python
plt.pie(df["smoking"].value_counts(),shadow=True, autopct='%1.1f%%',labels=['Non-smoker','Smoker'])
plt.show()
```

```python
df["smoking"].value_counts()
```

```python
plt.pie(df["gender"].value_counts(),shadow=True, autopct='%1.1f%%',labels=['male','female'])
plt.show()
```

```python
sns.displot(data=df,x='gender',kind='hist',col='smoking')
plt.show()
```

```python
sns.boxplot(data=df,x='age')
plt.show()
```

```python
sns.displot(data=df,x='age',kind='hist',col='gender',hue='smoking',bins=10,element='step')
plt.show()
```

```python
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(14,10))
ax[0,0].boxplot(df['height(cm)'])
ax[0,1].boxplot(df['weight(kg)'])

ax[1,0].hist(df['height(cm)'])
ax[1,1].hist(df['weight(kg)'])

ax[0,0].set_title("Height(cm)")
ax[0,0].set_ylabel('Height(cm)')
ax[1,0].set_xlabel('Height(cm)')

ax[0,1].set_title("Weight(kg)")
ax[0,1].set_ylabel('Weight(kg)')
ax[1,1].set_xlabel('Weight(kg)')
```

```python
sns.displot(data=df,x='gender',kind='hist',row='smoking',col='tartar',)
plt.show()
```

```python
sns.displot(data=df,x='gender',kind='hist',row='smoking',col='dental caries',)
plt.show()
```

```python
sns.catplot(data=df,y='hemoglobin',x='age', kind='bar',col='gender',hue='smoking',ci=None)
plt.show()
```

```python
sns.catplot(data=df,y='systolic',x='age', kind='bar',col='gender',hue='smoking',ci=None)
plt.show()
```

```python
sns.catplot(data=df,y='relaxation',x='age', kind='bar',col='gender',hue='smoking',ci=None)
plt.show()
```

```python
sns.catplot(data=df,y='fasting blood sugar',x='age', kind='bar',col='gender',hue='smoking',ci=None)
plt.show()
```
